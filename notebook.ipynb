```python
import matplotlib.pyplot as plt
import scipy.stats as stat
import pylab


def plot_sample(sample):
    
    plt.figure(figsize=(10,6))
    plt.subplot(1,2,1)
    sns.histplot(sample,kde=True)
    plt.subplot(1,2,2)
    stat.probplot(sample,dist='norm',plot=pylab)
    plt.show()
    
```

```python
plot_sample(log_normal_dist_data)
```

```python
plot_sample(np.log(log_normal_dist_data))
```

## uniform distribution


```python
from numpy import random 
sns.distplot(random.uniform(size=500),hist=True) #continous variable 
```

```python
sns.histplot(np.random.randint(1,8,size=800)) #discrete variable 
```

## bernoullli dist

```python
p=0.3
sample_size = 1000

bernoulli_data = np.random.choice([0,1], size=sample_size,p=[1-p,p])
sns.histplot(bernoulli_data,bins=2,kde=False,color='green')
```

## binomial dist

```python
n= 10
p=0.6

binomial_data=np.random.binomial(n,p,10000)
sns.distplot(binomial_data)
```

## poisson dist

```python
lamd = 5

poisson_data =  np.random.poisson(lamd,size=1000)

#plotting a histogram
ax= sns.distplot(poisson_data,kde=False,color='blue')
ax.set(xlabel='Poisson distribution',ylabel='Frequency')
plt.show()
```

lamd = 5 means the code is generating random data points from poisson dist where the avg number of evernts in the specified interval
